```python
# import pandas as pd
# import numpy as np
# from sklearn.model_selection import train_test_split, GridSearchCV
# from sklearn.preprocessing import StandardScaler, FunctionTransformer, OneHotEncoder
# from sklearn.pipeline import Pipeline
# from sklearn.compose import ColumnTransformer
# from xgboost import XGBRegressor
# from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, recall_score, f1_score

# data_path = 'dataset_with_labels_cleaned.csv'
# df = pd.read_csv(data_path)

# df = df.drop(["pr_count", "achievement_count", "name", "sport_type" ,"time_bins", "date", "time"], axis=1)

# X = df.loc[:, df.columns != 'performance_label']
# y = df['performance_label']

# # Ensure X is a DataFrame when passed to ColumnTransformer
# X = pd.DataFrame(X)

# sqrt_pipeline = Pipeline(
#     steps=[
#         ('sqrt_transform', FunctionTransformer(np.sqrt, feature_names_out='one-to-one')),
#         ('standard_scaler', StandardScaler())
#     ]
# )

# numeric_pipeline = Pipeline(
#     steps=[
#         ('standard_scaler', StandardScaler())
#     ]
# )

# preprocessor = ColumnTransformer(
#     transformers=[
#         ('sqrt_transform', sqrt_pipeline, ['kilojoules', 'total_elevation_gain', 'distance']),
#         ('nums', numeric_pipeline, ['average_heartrate', 'average_watts', 'moving_time', 'weighted_average_watts', 'average_cadence', 'average_speed'])
#     ], remainder='passthrough'
# )

# xgboost_pipeline = Pipeline(
#     steps=[
#         ('preprocess', preprocessor),
#         ('xgboost', XGBRegressor(objective='reg:squarederror', random_state=42))
#     ]
# )

# param_grid = {
#     'xgboost__n_estimators': [100, 300, 600, 1000],
#     'xgboost__learning_rate': [0.01, 0.1, 0.2, 0.3],
#     'xgboost__max_depth': [3, 5, 7, 9]
# }

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# grid_search = GridSearchCV(estimator=xgboost_pipeline, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='neg_root_mean_squared_error')
# grid_search.fit(X_train, y_train)

# best_params = grid_search.best_params_
# best_model = grid_search.best_estimator_

# print(f"Best parameters found: {best_params}")

# final_mse = mean_squared_error(y_test**2, grid_search.predict(X_test)**2, squared=False)
# final_mse


# X_test_transformed = best_model.named_steps['preprocess'].transform(X_test)

# # Prediction on test data
# y_pred = best_model.named_steps['xgboost'].predict(X_test_transformed)

# # Evaluate on test data
# final_rmse = mean_squared_error(y_test, y_pred, squared=False)
# print(f"Final RMSE on test data: {final_rmse}")


# raw_results = pd.DataFrame(
#     {
#         'Predicted Value': y_pred,
#         'True Value': y_test.reset_index(drop=True),
#     }
# ).assign(Difference=lambda x: x['Predicted Value'] - x['True Value'])

# # Display the results
# print(raw_results)

# # Convert regression predictions to binary classification
# threshold = 0.5
# y_test_class = (y_test > threshold).astype(int)
# y_pred_class = (y_pred > threshold).astype(int)

# # Compute confusion matrix
# conf_matrix = confusion_matrix(y_test_class, y_pred_class)
# print(f"Confusion Matrix:\n{conf_matrix}")

# # Compute recall
# recall = recall_score(y_test_class, y_pred_class)
# print(f"Recall: {recall}")

# # Compute F1 score
# f1 = f1_score(y_test_class, y_pred_class)
# print(f"F1 Score: {f1}")











import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from xgboost import XGBRegressor
import matplotlib.pyplot as plt

# Load the dataset
data_path = 'dataset_with_labels_cleaned.csv'  # Adjust the path to your dataset
df = pd.read_csv(data_path)

# Drop unnecessary non-numeric columns
df = df.drop(["pr_count", "achievement_count", "name", "sport_type" ,"time_bins", "date", "time"], axis=1)

# Split the dataset into features (X) and target (y)
X = df.loc[:, df.columns != 'performance_label']
y = df['performance_label']

# Ensure X is a DataFrame when passed to ColumnTransformer
X = pd.DataFrame(X)

# Define the pipelines for numeric and square root transformation
sqrt_pipeline = Pipeline(
    steps=[
        ('sqrt_transform', FunctionTransformer(np.sqrt, feature_names_out='one-to-one')),
        ('standard_scaler', StandardScaler())
    ]
)

numeric_pipeline = Pipeline(
    steps=[
        ('standard_scaler', StandardScaler())
    ]
)

# Define the preprocessor with ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('sqrt_transform', sqrt_pipeline, ['kilojoules', 'total_elevation_gain', 'distance']),
        ('nums', numeric_pipeline, ['average_heartrate', 'average_watts', 'moving_time', 'weighted_average_watts', 'average_cadence', 'average_speed'])
    ], remainder='passthrough'
)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)

# Preprocess the training and test data
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)

# Train the XGBoost model with evaluation metrics
eval_set = [(X_train_transformed, y_train), (X_test_transformed, y_test)]
xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42, n_estimators=1000, learning_rate=0.1, max_depth=5)

xgb_model.fit(X_train_transformed, y_train, eval_metric='rmse', eval_set=eval_set, verbose=True, early_stopping_rounds=50)

# Extract training and validation metrics
results = xgb_model.evals_result()

# Plot learning curves
epochs = len(results['validation_0']['rmse'])
x_axis = range(0, epochs)

fig, ax = plt.subplots()
ax.plot(x_axis, results['validation_0']['rmse'], label='Train')
ax.plot(x_axis, results['validation_1']['rmse'], label='Test')
ax.legend()
plt.ylabel('RMSE')
plt.title('XGBoost RMSE over Epochs')
plt.show()

```
